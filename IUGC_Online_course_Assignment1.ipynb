```python
#Exercise 1
#Waktu yang sudah berlalu semenjak Ulang tahun anda
from obspy.core import UTCDateTime
time1=UTCDateTime() #memanggil UTCTime sekarang
time2=UTCDateTime(2020,5,27)#ulang tahun saya
print('Waktu saat ini: ', time1)
print('Ulang tahun saya: ',time2)

print ('Waktu yang sudah berlalu sejak ulang tahun saya: ',time1-time2, 'detik')
```

```python
#Exercise 3
#Plot waveform dengan panjang data hanya 10 menit yang mencakup noise dan gelombang P 
#Gunakan atribut starttime dan endtime pada .plot()

from obspy.core import read
from obspy.core import UTCDateTime
st = read('./Data/IU_ULN_2015-07-18T02.mseed')

print (st)

singlechannel=st[0]
singlechannel.plot()
time=UTCDateTime(2015,7,18,2,45)
singlechannel.plot(color='white', bgcolor='navy', number_of_ticks=3, starttime=time, endtime=time+600)
```
